## 1.1 Create a two-dimensional vector and plot it on a graph

## 1.2 Create a three-dimensional vector and plot it on a graph

```
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

origin = (0,0,0)
endpoint = (0.7,0.9,0.4)
vector = np.array([*origin, *endpoint])
# vector = zip(*vector) # turn the array sideways

fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.quiver(*vector, length=1)
ax.set(xlim=[0, 1], ylim=[0, 1], zlim=[0, 1])
ax.set(xlabel='X', ylabel='Y', zlabel='Z')
plt.show()
```

## 1.3 Scale the vectors you created in 1.1 by $5$, $\pi$, and $-e$ and plot all four vectors (original + 3 scaled vectors) on a graph. What do you notice about these vectors?

## 1.8 Find $||g||$ and then find $||h||$. Which is longer?

\begin{align}
\vec{g} = \begin{bmatrix} 1 \\ 1 \\ 1 \\ 8 \end{bmatrix}
\qquad
\vec{h} = \begin{bmatrix} 3 \\3 \\ 3 \\ 3 \end{bmatrix}
\end{align}

```
from numpy import linalg
import math
g = np.array([1,1,1,8])
h = np.array([3,3,3,3])
norm_g = linalg.norm(g)
norm_h = linalg.norm(h)
print('||g|| =',norm_g)
print('||h|| =',norm_h)

if norm_g > norm_h:
  print('||g|| is longer than ||h||.')
elif norm_h > norm_g:
  print('||h|| is longer than ||g||.')
else:
  print('||g|| and ||h|| are equal.')
```

\qquad
D = \begin{bmatrix}

0 & 0 & 1
\end{bmatrix}
\qquad
E = \begin{bmatrix}
1 & 3 \\
5 & 7
\end{bmatrix}
\end{align}

```
import numpy as np

A = np.array([[1,2],
              [3,4],
              [5,6]])

B = np.array([2,4,6])

C = np.array([[9,6,3],
              [4,7,11]])

D = np.array([[1,0,0],
              [0,1,0],
              [0,0,1]])

E = np.array([[1,3],
              [5,7]])

matrices = [A,B,C,D,E]
names = ['A','B','C','D','E']

for i, first in enumerate(matrices):
  for j, second in enumerate(matrices):
    print(names[i],'x',names[j],': ',first.shape,'x',second.shape)
    try:
      product = np.matmul(first,second)
      print(first,'x',second,'=',product)
    except:
      print('cannot be multiplied')
```

- If a matrix has a shape of (3,), interpret this as (1,3) for multiplication purposes.
- Two matrices can only be multiplied if the two inner dimensions match.
- The resulting shape will be the two outer dimensions.
